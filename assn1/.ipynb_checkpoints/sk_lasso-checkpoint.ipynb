```python
import numpy as np
from sklearn import linear_model
```

```python
Z = np.loadtxt( "train" )

y = Z[:,0]
X = Z[:,1:]
```

```python
lamda = 0.01
```

```python
clf = linear_model.Lasso(alpha=0.01)
```

```python
clf.fit(X,y)
clf
```

```python
clf.coef_
```

```python
c = 0
for i in clf.coef_:
    if i != 0:
        c = c + 1
print(c)
```

```python
def f(x,w,lamda):
    t = (x.dot(w)-y).T*(x.dot(w)-y)
    return np.sum(t) + lamda*np.sum(abs(x))
f(X,w,lamda)
```
